```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
d = sns.load_dataset('iris')
```

```python
d.head()
```

Now let's first plot a regular scatter plot => [sepal_width vs. sepal_length] without any regression plotting.

```python
sns.regplot(x = d['sepal_length'], y = d['sepal_width'], fit_reg = False)
plt.show()
```

It's said "A picture speaks thousand words". We had seen the head of the dataset in the earlier cell. However this scatterplot gives a lot clearer view of the same. We can also quickly check the ouliers in the data.

Now lets plot the scatter with *regression*
# Regression

```python
sns.regplot(x = d['sepal_length'], y = d['sepal_width'])
plt.show()
```

The linear regression doesn't fit the data properly. We can see there are numorous points away from the regression line.
Let's try increasing the order.

# Order - 2

```python
sns.regplot(x = d['sepal_length'], y = d['sepal_width'], order = 2)
plt.show()
```

# Order - 3

```python
sns.regplot(x = d['sepal_length'], y = d['sepal_width'], order = 3)
plt.show()
```

# Order - 4

```python
sns.regplot(x = d['sepal_length'], y = d['sepal_width'], order = 4)
plt.show()
```

Now wouldn't it be better to distribute the categories of flower so that they are clearly visible in the scatterplot.
So let's do it then.

```python
sns.lmplot(x = 'sepal_length', y = 'sepal_width', data = d, fit_reg = False, hue = 'species')
plt.show()
```

Now let's venture into a new kind of visualization. 
# Marginal Histograms
Here we plot histograms on the axis to get a more easy view on the density of the scatter plots along the axis.

```python
sns.jointplot(x = d['sepal_length'], y = d['sepal_width'], kind = 'scatter')
plt.show()
```

Here, for example, we can see the density of points is higher near y = 3.0 and x = 6.0. We also can see the outliers with the help of the histograms plotted!
